```python
#Sharpning Images
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('Nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gass=cv2.GaussianBlur(img,(5,5),0)
median=cv2.medianBlur(img,ksize=5)
sharp=cv2.addWeighted(img,1.65,gass,-0.68,0)
sharp1=cv2.addWeighted(img,1.60,median,-0.55,40)

plt.figure(figsize=(10,6))
plt.subplot(1,3,1),plt.imshow(img),plt.axis('off'),plt.title('Source Image')
plt.subplot(1,3,2),plt.imshow(sharp),plt.axis('off'),plt.title('Sharped image')
plt.subplot(1,3,3),plt.imshow(sharp1),plt.axis('off'),plt.title('Sharped image')

```
