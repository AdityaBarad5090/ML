```python
#Applying Kernal 2D to Image

import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('1.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
k1=np.array([[0,0,0],[0,2,0],[0,0,0]])
k2=np.array([[0,0,0],[0,3,-1],[0,0,0]])
k1output=cv2.filter2D(img,-1,k1)
k2output=cv2.filter2D(img,-1,k2)
plt.subplot(1,3,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(k1output),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(k2output),plt.axis('off')
```

```python
#Blur the Image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('1.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
k1=np.ones((4,4),np.float32)/16
k1output=cv2.filter2D(img,-1,k1)
plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(k1output),plt.axis('off')
```
