```python
#Edge Detection With Sobel

import matplotlib.pyplot as plt
import cv2

img=cv2.imread('tiger.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,50)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,50)
sobel=cv2.magnitude(sobelx,sobely)

plt.subplot(2,2,1),plt.axis('off'),plt.imshow(img)
plt.subplot(2,2,2),plt.axis('off'),plt.imshow(sobelx,cmap='gray')
plt.subplot(2,2,3),plt.axis('off'),plt.imshow(sobely,cmap='gray')
plt.subplot(2,2,4),plt.axis('off'),plt.imshow(sobel,cmap='gray')
```

```python
#Edge Detection With Laplacian

import matplotlib.pyplot as plt
import cv2

img=cv2.imread('tiger.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gaussian=cv2.GaussianBlur(img,(5,5),0)

laplacian=cv2.Laplacian(img,cv2.CV_64F,ksize=5)

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(laplacian,cmap='gray'),plt.axis('off')
#plt.subplot(1,3,3),plt.imshow(gaussian,cmap='gray'),plt.axis('off')


```

```python
#Edge Detection with Canny

import matplotlib.pyplot as plt
import cv2

img=cv2.imread('tiger.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gaussian=cv2.GaussianBlur(img,(5,5),0)

canny=cv2.Canny(img,threshold1=10,threshold2=100)

plt.subplot(1,2,2),plt.imshow(canny,cmap='gray'),plt.axis('off')
plt.subplot(1,2,1),plt.imshow(gaussian),plt.axis('off')
```
