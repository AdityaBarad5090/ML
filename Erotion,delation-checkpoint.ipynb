```python
#Erotion and Dilation
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('one.png',0)

kernal = np.ones((3,3),np.uint8)

#Apply an Erosation
erosioan = cv2.erode(img,kernal,iterations=1)

delt = cv2.dilate(img,kernal,iterations=1)

plt.subplot(1,3,1),plt.imshow(img,cmap='gray'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(erosioan,cmap='gray'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(delt,cmap='gray'),plt.axis('off')
```

```python
img=cv2.imread('flow4.png')

ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,cr,cb=cv2.split(ycrcb)

plt.subplot(2,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.axis('off'),plt.title('Source')
plt.subplot(2,2,2),plt.imshow(y,cmap='gray'),plt.axis('off'),plt.title('Y (Luminance)')
plt.subplot(2,2,3),plt.imshow(cr,cmap='gray'),plt.axis('off'),plt.title('cr (crominance Red)')
plt.subplot(2,2,4),plt.imshow(cb,cmap='gray'),plt.axis('off'),plt.title('cb (crominance Blue)')
```
