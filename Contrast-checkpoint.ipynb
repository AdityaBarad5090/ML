```python
#Contrast Apply To Inage
import cv2
import matplotlib.pyplot as plt

img= cv2.imread('flow2.jpg',0)
img1=cv2.equalizeHist(img)

plt.subplot(1,2,1),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Source')
plt.subplot(1,2,2),plt.axis('off'),plt.imshow(img1,cmap='gray'),plt.title('Contrast')
```

```python
#Contrast Applt to Color image
img = cv2.imread('flow3.jpg')

ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,cr,cb=cv2.split(ycrcb)

y_eq=cv2.equalizeHist(y)
ycrcb_eq=cv2.merge([y_eq,cr,cb])
glob_eq=cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off'),plt.title('Source')
plt.subplot(1,2,2),plt.imshow(glob_eq),plt.axis('off'),plt.title('Contrsat')
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe = clahe.apply(y)

ycrcb_clahe = cv2.merge([y_clahe,cr,cb])

clahe_eq = cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2BGR)
plt.figure(figsize=(7,5))

plt.subplot(1,2,1),plt.imshow(img),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(clahe_eq),plt.axis('off')
```
